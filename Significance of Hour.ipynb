```python
import pandas as pd
import numpy as np

import scipy.stats as st
```

### DMA - Import .xlsx of the file

```python
hour_path = f"Resources/BR _ Revlimid - HCP - Hour.xlsx"
hour_df = pd.read_excel(hour_path)

hour_df["Hour of day"].fillna("None", inplace=True)
hour_df.drop(columns=["Row Type","Avg pos"], inplace=True)
hour_df.sort_values("Cost", ascending=False)

hour_df= hour_df.dropna(0, how="any")

hour_df.sort_values("Hour of day")
```

```python
# get the means
hour_CWTA_mean = hour_df["Cost/wt action"].mean()
print("Cost per Weighted Total Action in hour of day: $"+str(round(hour_CWTA_mean, 2)))

hour_click_mean = hour_df["Clicks"].mean()
print("Mean of number of clicks in hour of day: "+ str(round(hour_click_mean, 2) ))

```

```python
#Chi square test, to see if the cost/weighted average is truly different than one another

print(st.chisquare(hour_df["Cost/wt action"]))
#np.warnings.filterwarnings('ignore')

print("If PValue is bigger than 0.05, that means the difference of CWTA between hours are pure chance")
```
